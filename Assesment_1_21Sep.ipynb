```python
student_lst = [
    "Reeve Ewer",
    "Tex Ryta",
    "Theresa Green",
    "Barry Kade",
    "Stan Dupp",
    "Neil Down",
    "Con Trariweis",
    "Don Messwidme",
    "Al Annon",
    "Anna Domino",
]

roll_lst = [
    "Reeve Ewer-200",
    "Tex Ryta-100",
    "Theresa Green-125",
    "Barry Kade-225",
    "Stan Dupp-365",
    "Neil Down-451",
    "Con Trariweis-114",
    "Don Messwidme-335",
    "Al Annon-455",
    "Anna Domino-136",
]
```

```python
dict1 = {}
for i in roll_lst:
    parts = i.split('-')
    if len(parts) == 2:
        student_name, roll_number = parts
        dict1[student_name] = int(roll_number)
print(dict1)
```

```python
def key1(name):
    return dict1.get(name)

sorted_student_names = sorted(student_lst, key=key1)

print(sorted_student_names)
```

```python
print("Sorted Student Names and roll Numbers:\n")
for name in sorted_student_names:
    print(f"Name: {name}, Roll Number: {dict1.get(name)}")
```
